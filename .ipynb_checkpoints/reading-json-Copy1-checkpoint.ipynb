## JSON as a String (Raw Scraped data)

```python
files = ['resume_output_datascientist_california.json', 
        'resume_output_datascientist_texas.json',
        'resume_output_datascientist_nc.json',
        'resume_output_datascientist_chicago.json',
        'resume_output_datascientist_newyork.json']

rf = files[0]
```

```python
with open(rf, 'r+') as f:
    f.seek(0,0)
    f.write('[')
```

```python
with open(rf, 'rb+') as f:
    f.seek(-1,os.SEEK_END)
    f.truncate()
```

```python
with open(rf, 'a') as f:
    f.write(']')
```

```python
data = []
with open(rf) as f:
    s = f.read()
    print(s)
```

```python
type(s)
```

```python
# s
```

```python
s = s.replace('\\n', '')
s = s.replace('\\', '')
s = s.replace('}""{', '},\n{')
```

```python
outfile = rf[:-5]+'_semi_processed.json'
with open(outfile, 'w') as f:
    f.write(s)
```

```python
'''This may give you the following error:
JSONDecodeError: Expecting ',' delimiter: line ## column ### (char #####)

For now, manually go to that line/column and 
explicitly assign that as a literal caracter by adding "\"

We can programmatically add that later
'''
with open(outfile) as f:
    data = json.load(f)
```
